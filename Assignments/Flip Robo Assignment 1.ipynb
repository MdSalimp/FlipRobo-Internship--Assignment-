# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.¶

```python
/#page url
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
soup=BeautifulSoup(page.content)
page
# Scrap Movie name, Year , rating
movie_name=[i.text.split('\n')[2]  for i in soup.find_all( class_="lister-item-header")]
year=[i.text.split('\n')[3] for i in soup.find_all( class_="lister-item-header")]
rating=[i.text.strip()for i in soup.find_all(class_="inline-block ratings-imdb-rating")]
df=pd.DataFrame({'Name':movie_name,'Rating':rating,'Year':year})
df
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup=BeautifulSoup(page.content)
team = soup.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = [] #empty list
points = [] #empty list
ratings = [] #empty list
new_list = [] #empty list

for i in soup.find_all("td",class_='rankings-block__banner--matches'): # first place team number of matches
    matches.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--points'):# first place team points
    points.append(i.text)
for i in soup.find_all("td",class_='rankings-block__banner--rating u-text-right'):# first place team ratings
    ratings.append(i.text.replace("\n",""))
for i in soup.find_all("td",class_='table-body__cell u-center-text'):# other teams number of matches and points
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i]) # other teams matches
    points.append(new_list[i+1]) # other teams points
for i in soup.find_all("td",class_='table-body__cell u-text-right rating'):# other teams ratings
    ratings.append(i.text)
    
# Make data frame of top 10 ICC teams
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc    
```

7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :

```python
News_soup=BeautifulSoup(requests.get("https://www.cnbc.com/world/?region=world").content,"html.parser")
Headline=[i.text for i in News_soup.find_all(class_="LatestNews-headline") ] # Scrap Headlines
Time=[i.text for i in News_soup.find_all("span",class_="LatestNews-wrapper") ] # Scrap Time
# Scrap news Links
news = News_soup.find_all('div', {'class','LatestNews-container'})
lnk = [each.a.get('href') for each in news]
# Create data frame
news1=pd.DataFrame({'Headlines':Headline,'Time':Time,"Link":lnk})
news1
```

8) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.

```python
scholar_soup=BeautifulSoup(requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en").content,"html.parser")
rank=[i.text for i in scholar_soup.find_all("td",class_="gsc_mvt_p") ] # scrap rank
publication=[i.text for i in scholar_soup.find_all("td",class_="gsc_mvt_t") ] # scrap publication
index=[i.text for i in scholar_soup.find_all("td",class_="gsc_mvt_n") ] # scrap index
h5_index=[index[i] for i in range(0,len(index),2)] # split h5 index
h5_median=[index[i] for i in range(1,len(index),2)] # split h5-median
df=pd.DataFrame({"Rank":rank,"Publication":publication,"h5-index":h5_index,"h5-median":h5_median})
df
```

10) Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en

i) Rank

ii) Publication

iii) h5-index

iv) h5-median

```python
#send get request to the webpage server to get the source code of the page and download the page content
scholer = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
scholer
```

```python
conn = BeautifulSoup(scholer.content)
publication = []
for i in conn.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)

h5index = []
for i in conn.find_all('a',class_="gsc_mvt_n"):
    h5index.append(i.text)

h5median = []
for i in conn.find_all('span',class_="gsc_mp_anchor gsc_mp_tgh"):
    h5median.append(i.text)
rank = []
for i in range(1,101):
    rank.append(i)

metrics = pd.DataFrame({'Publication':publication, 'h5-index':h5index,'h5-median':h5median}, index = rank)
metrics
```
